```python
import matplotlib.pyplot as plt
import pickle

def get_hists(exp_names):
    hists = {}
    out_dir = 'output'
    for exp_name in exp_names:
        history_file = f"{out_dir}/{exp_name}/history.pickle"
        with open(history_file, 'rb') as history_obj:
            hist = pickle.load(history_obj)
        hists[exp_name] = hist
    return hists

def plot_stat(hists, metric):
    for exp_name, hist in hists.items():
        plt.plot(hist[metric], label=exp_name)
    plt.title(metric)
    plt.legend()
    plt.show()
```

```python
import os
hists = get_hists(os.listdir('output'))
plot_stat(hists, 'val_accuracy')
plot_stat(hists, 'val_loss')
plot_stat(hists, 'accuracy')
plot_stat(hists, 'loss')
```
